## 4.1 Teorema Probabilitas Total

```python
#Kode Praktik - Python
# Data
total_population = 900
cancer_smoker = 45
cancer_nonsmoker = 20

# Hitung total penderita kanker
total_cancer = cancer_smoker + cancer_nonsmoker

# Probabilitas seseorang yang terpilih menderita kanker paru-paru
prob_cancer = total_cancer / total_population

# Output
print("Total penderita kanker:", total_cancer)
print("Probabilitas seseorang yang dipilih menderita kanker paru-paru: {:.4f}".format(prob_cancer))
#Output - Python

#Total penderita kanker: 65
#Probabilitas seseorang yang dipilih menderita kanker paru-paru: 0.0722
```

## 4.2 Probabilitas Kejadian Bebas dan Bersyarat

```python
#Kode Praktik - Python
# Probabilitas A hidup 50 tahun
P_A = 0.7

# Probabilitas B hidup 90 tahun
P_B = 0.1

# Asumsi A dan B independen
P_A_and_B = P_A * P_B
#Output - Python

#Probabilitas bahwa A dan B hidup dalam 90 tahun adalah: 0.07

```

```python
#Kode Praktik - Python
# Jumlah apel
apel_hijau = 10
apel_merah = 15
total_apel = apel_hijau + apel_merah

# P(B): peluang apel merah diambil pertama
P_B = apel_merah / total_apel

# P(A|B): peluang apel hijau diambil kedua, setelah merah diambil
P_A_given_B = apel_hijau / (total_apel - 1)

# P(A ∩ B): merah diambil dulu, lalu hijau
P_A_and_B = P_B * P_A_given_B
#Output - Python
print(f"P(A ∩ B) = {P_A_and_B:.4f}")
```

```python
#Kode Praktik - Python
# Data jumlah mahasiswa
diterima_laki = 25
ditolak_laki = 150
total_laki = diterima_laki + ditolak_laki

diterima_perempuan = 40
ditolak_perempuan = 200
total_perempuan = diterima_perempuan + ditolak_perempuan

total_seluruh = total_laki + total_perempuan

# P(A|B): Probabilitas diterima jika laki-laki
P_A_given_B = (diterima_laki / total_seluruh) / (total_laki / total_seluruh)

# P(A|B^c): Probabilitas diterima jika perempuan
P_A_given_Bc = (diterima_perempuan / total_seluruh) / (total_perempuan / total_seluruh)

# P(A^c|B): Probabilitas ditolak jika laki-laki
P_Ac_given_B = (ditolak_laki / total_seluruh) / (total_laki / total_seluruh)
#Output - Python
# Cetak hasil
print("P(A|B) =", P_A_given_B)    # 1/7
print("P(A|Bᶜ) =", P_A_given_Bc)  # 1/6
print("P(Aᶜ|B) =", P_Ac_given_B)  # 6/7

```

##4.3	 Teorema Bayes

```python
#Kode Praktik - Python
#Probabilitas pelanggan ditempatkan di hotel
P_I = 0.2
P_B = 0.5
P_S = 0.3

# Probabilitas kamar mandi rusak di tiap hotel
P_D_given_I = 0.05
P_D_given_B = 0.04
P_D_given_S = 0.08

# 1. Probabilitas total kamar mandi rusak (P(D))
P_D = (P_I * P_D_given_I) + (P_B * P_D_given_B) + (P_S * P_D_given_S)

# 2. Probabilitas pelanggan menginap di Hotel S jika kamar mandi rusak (P(S|D)) - Teorema Bayes
P_S_given_D = (P_S * P_D_given_S) / P_D
#Output - Python
# Output
print(f"Peluang kamar mandi tidak baik (P(D)) = {P_D:.4f}")
print(f"Peluang pelanggan di Hotel S jika kamar mandi tidak baik (P(S|D)) = {P_S_given_D:.4f}")

#output:
#Peluang kamar mandi tidak baik (P(D)) = 0.0540
#Peluang pelanggan di Hotel S jika kamar mandi tidak baik (P(S|D)) = 0.4444
```
