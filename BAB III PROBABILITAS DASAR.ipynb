## Aturan Penjumlahan

```python
#Kode Praktik – Python
# Jumlah jenis pilihan
jumlah_kue = 3
jumlah_minuman = 4
# Aturan penjumlahan (jika hanya memilih salah satu)
total_pilihan = jumlah_kue + jumlah_minuman
# Cetak hasil
print("Total kemungkinan pilihan:", total_pilihan)
#Output– Python
#Total kemungkinan pilihan:7
```

## Aturan Perkalian

```python
#Kode Praktik – Python
# Jumlah pilihan di setiap tahap
kue = 3
minuman = 4
# Hitung total kombinasi
total_kombinasi = kue * minuman
#Output – Python
# Cetak hasil
print(total_kombinasi)  # Output: 12
```

## Kombinasi

```python
#Kode Praktik – Python
import math
n = 5
r = 3
kombinasi = math.comb(n, r)
print("Jumlah kombinasi:", kombinasi)
#Output Python:
#Jumlah kombinasi: 10
```

## Permutasi

```python
#Kode Praktik – Python
import math
n = 5
r = 3
permutasi = math.perm(n, r)
print("Jumlah permutasi:", permutasi)
#Output Python:
#Jumlah permutasi: 60
```

## Probabilitas Dasar

```python
#Kode Praktik – Python
jumlah_keinginan = 3  # Angka genap: 2, 4, 6
jumlah_total = 6
probabilitas = jumlah_keinginan / jumlah_total
print("Probabilitas muncul angka genap:", probabilitas)
#Output Python:
#Probabilitas muncul angka genap: 0.5
```

###a.Kejadian Saling Lepas (Mutually Exclusive Events)

```python
#Kode Praktik – Python
P_A = 1/6
P_B = 1/6
P_union = P_A + P_B
print("Probabilitas A atau B terjadi:", P_union)
#Output Python:
#Probabilitas A atau B terjadi: 0.3333333333333333
```

###b.b.	Kejadian Tidak Saling Lepas (Non-Mutually Exclusive Events)

```python
#Kode Praktik – Python
P_A = 0.6
P_B = 0.4
P_A_dan_B = 0.25
P_union = P_A + P_B - P_A_dan_B
print("Probabilitas A atau B terjadi:", P_union)
#Output Python:
#Probabilitas A atau B terjadi: 0.75
```

```python
Kode Praktik – Python
P_A = 1/6  # Probabilitas muncul angka 2 pada dadu
P_B = 1/2  # Probabilitas muncul gambar pada koin
P_intersection = P_A * P_B
print("Probabilitas A dan B terjadi bersamaan:", P_intersection)
Output Python:
Probabilitas A dan B terjadi bersamaan: 0.08333333333333333
```

# Aturan Perkalian Probabilitas
##a.	Kejadian Independen (Independent Events)

```python
#Kode Praktik – Python
P_A = 1/6  # Probabilitas muncul angka 2 pada dadu
P_B = 1/2  # Probabilitas muncul gambar pada koin
P_intersection = P_A * P_B
print("Probabilitas A dan B terjadi bersamaan:", P_intersection)
##Output Python:
#Probabilitas A dan B terjadi bersamaan: 0.08333333333333333
```

```python
#Kode Praktik – Python
P_A = 5/8           # Probabilitas bola pertama merah
P_B_given_A = 4/7   # Probabilitas bola kedua merah jika bola pertama merah
P_intersection = P_A * P_B_given_A
print("Probabilitas kedua bola merah:", P_intersection)
##Output Python:
#Probabilitas kedua bola merah: 0.35714285714285715
```

##b.	Kejadian Tidak Independen (Dependent Events)

```python
#Kode Praktik – Python
P_A = 5/8           # Probabilitas bola pertama merah
P_B_given_A = 4/7   # Probabilitas bola kedua merah jika bola pertama merah
P_intersection = P_A * P_B_given_A
print("Probabilitas kedua bola merah:", P_intersection)
#Output Python:
#Probabilitas kedua bola merah: 0.35714285714285715
```
