#5.3	Distribusi Probabilitas Diskrit
##5.3.1	Fungsi Massa Probabilitas (probability mass function / PMF)

```python
#Kode Praktik - Python
# Fungsi faktorial manual
def faktorial(n):
    hasil = 1
    for i in range(1, n+1):
        hasil *= i
    return hasil
# Fungsi kombinasi manual
def kombinasi(n, k):
    return faktorial(n) // (faktorial(k) * faktorial(n - k))
# Hitung probabilitas
total = kombinasi(20, 2)

P0 = kombinasi(3, 0) * kombinasi(17, 2) / total
P1 = kombinasi(3, 1) * kombinasi(17, 1) / total
P2 = kombinasi(3, 2) * kombinasi(17, 0) / total
# Cetak hasil
print(f"P(X = 0) = {P0:.4f}")
print(f"P(X = 1) = {P1:.4f}")
print(f"P(X = 2) = {P2:.4f}")
print(f"Total Probabilitas = {P0 + P1 + P2:.4f}")

# Output Program Python dan R:
# P(X = 0) = 0.7158
# P(X = 1) = 0.2684
# P(X = 2) = 0.0158
# Total Probabilitas = 1.0000
```

##5.3.2	Fungsi Distribusi Kumulatif (CDF)

```python
#Kode Praktik- Python

from math import comb
N, K, n = 20, 3, 2  # total, rusak, dibeli
total = comb(N, n)
p0 = comb(N-K, n) / total                 # 2 baik
p1 = comb(K, 1) * comb(N-K, 1) / total    # 1 rusak, 1 baik
p2 = comb(K, 2) / total                   # 2 rusak
cdf_leq1 = p0 + p1
print(f"P(X=0)   = {p0:.6f}")
print(f"P(X=1)   = {p1:.6f}")
print(f"P(X=2)   = {p2:.6f}")
print(f"P(X<=1)  = {cdf_leq1:.6f}")

# Output
# P(X=0)   = 0.715789
# P(X=1)   = 0.268421
# P(X=2)   = 0.015789
# P(X<=1)  = 0.984211

```

##5.3.3	Rata-rata (Mean) dan Varians dari Distribusi Probabilitas Diskrit

```python
#Kode Praktik - Python
x = [0, 1, 2]
pmf = [0.7158, 0.2684, 0.0158]
mean = sum(xi * pi for xi, pi in zip(x, pmf))
ex2 = sum((xi**2) * pi for xi, pi in zip(x, pmf))
variance = ex2 - mean**2
print(f"Mean: {mean:.4f}")
print(f"Variance: {variance:.4f}")
# Output - Python
# Mean: 0.3000
# Variance: 0.2416
```

```python
#Code Praktik – Python
import numpy as np
# Data
x = np.array([50, 100, 200])
p = np.array([0.2, 0.5, 0.3])
# Mean
mean = np.sum(x * p)
# Variance
variance = np.sum((x - mean)**2 * p)
# Std Dev
std_dev = np.sqrt(variance)
print(f"Mean: {mean}")
print(f"Variance: {variance}")
print(f"Standard Deviation: {std_dev}")
# Output - PYTHON
# Mean: 120.0
# Variance: 3100.0
# Standard Deviation: 55.67764362830022
```

```python
#Kode Praktik- Python
import sympy as sp

x = sp.symbols('x')
f = 60 * sp.exp(-60*(30 - x))

# a) P(X < 29.94)
p_a = sp.integrate(f, (x, 29.9, 29.94))

# b) P(29.95 <= X <= 29.99)
p_b = sp.integrate(f, (x, 29.95, 29.99))

print("Proporsi gagal inspeksi (<29.94 mm):", float(p_a))
print("Proporsi 29.95 - 29.99 mm:", float(p_b))
```

##5.4.2	Fungsi Distirbusi Kumulatif (Cummulative Density Function)

```python
#Kode Praktik – Ptyhon
# Integrasi numerik pakai aturan trapesium (tanpa library)
def f(x):
    return 2 * (1 - x)
def trapezoid_rule(a, b, n):
    h = (b - a) / n
    total = f(a) + f(b)
    for i in range(1, n):
        total += 2 * f(a + i * h)
    return (h / 2) * total

# (a) Total probability from 0 to 1
p_total = trapezoid_rule(0, 1, 1000)
print("P(0 < X < 1) ≈", round(p_total, 4))

# (b) Probability from 1/4 to 1/2
p_interval = trapezoid_rule(0.25, 0.5, 1000)
print("P(1/4 < X < 1/2) ≈", round(p_interval, 4))

# OUTPUT - Python
# P(0 < X < 1) ≈ 1.0
# P(1/4 < X < 1/2) ≈ 0.3125
```
